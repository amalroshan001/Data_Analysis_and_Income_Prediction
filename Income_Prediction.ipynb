## Encoding Data

```python
from sklearn import preprocessing
```

```python
label_enc = preprocessing.LabelEncoder()
```

```python
df1 = df.copy()
```

```python
for col in df1:
    df1[col]=label_enc.fit_transform(df1[col].values)
```

```python
df1
```

## SMOTE of Data Sets

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE()

# fit predictor and target variable
x_smote , y_smote = smote.fit_resample(df1.iloc[:,0:-1],df1['Income'])

print('Original Dataset Shape : ', len(df1))
print('Resampled Dataset Shape : ', len(y_smote))
```

### Splitting The Data for Testing and Training

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_smote,y_smote,test_size=0.25,random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logicreg = LogisticRegression()
```

```python
logicreg.fit(x_train,y_train)
```

```python
y_predict = logicreg.predict(x_test)
```

```python
y_predict
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix
```

```python
acc = accuracy_score(y_test,y_predict)
```

```python
acc
```

```python
con = confusion_matrix(y_test,y_predict)
```

```python
con
```

```python
f1 = f1_score(y_test,y_predict)
```
